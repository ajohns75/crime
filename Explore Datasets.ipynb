## Find the percentage of null/na values in each dataframe

```python
# function to determine the percentage of null/na values in the DataFrame
def na_perc(df):
    df = df.isna().mean().mul(100).sort_values(ascending=False)
    return df
```

# Website I found while searching for definition of violent vs non violent crimes

[https://www.legalmatch.com/law-library/article/non-violent-vs-violent-crimes.html](https://www.legalmatch.com/law-library/article/non-violent-vs-violent-crimes.html)


[https://www.fightforyourrightsmn.com/blog/2016/12/the-difference-between-violent-and-non-violent-crimes.shtml](https://www.fightforyourrightsmn.com/blog/2016/12/the-difference-between-violent-and-non-violent-crimes.shtml)


[https://www.usf.edu/student-affairs/victim-advocacy/types-of-crimes/violent-crimes.aspx](https://www.usf.edu/student-affairs/victim-advocacy/types-of-crimes/violent-crimes.aspx)

```python
# making violent crime lists
balt_v = ['AGG. ASSAULT', 'COMMON ASSAULT', 'HOMICIDE', 'RAPE', 'SHOOTING']
mont_v = ['Forcible Sodomy','Fondling','Simple Assault','Sexual Assault With An Object',
          'Forcible Rape','Aggravated Assault','Forcible Fondling','Murder and Nonnegligent Manslaughter',
          'Kidnapping/Abduction','Justifiable Homicide','Negligent Manslaughter',
         'Human Trafficking, Commercial Sex Acts']
```

```python
# create columns in dataframes for violent [1] and nonviolent [0]
baltimore['violent'] = baltimore.apply(lambda x: int(x['Description'] in balt_v), axis=1)
mont['violent'] = mont.apply(lambda x: int(x['Crime Name2'] in mont_v), axis=1)
```
